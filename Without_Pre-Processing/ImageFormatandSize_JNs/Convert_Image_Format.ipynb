```python
import os
from PIL import Image

def convert_images_to_png(source_folder, target_folder):
    """
    Convert all images in the source folder and its subfolders to .png format and
    save them in a target folder with the same subfolder structure.
    
    :param source_folder: The root folder containing subfolders with images.
    :param target_folder: The root folder where converted images will be saved.
    """
    for subdir, dirs, files in os.walk(source_folder):
        for file in files:
            # Check if the file is an image
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff', '.bmp', '.gif')):
                file_path = os.path.join(subdir, file)
                # Calculate relative path to keep the directory structure
                relative_path = os.path.relpath(subdir, source_folder)
                target_dir = os.path.join(target_folder, relative_path)
                # Create the target directory if it does not exist
                if not os.path.exists(target_dir):
                    os.makedirs(target_dir)
                # Attempt to open the image file
                try:
                    with Image.open(file_path) as img:
                        # Define the new file name, replacing the original extension with .png
                        new_file_name = os.path.splitext(file)[0] + '.png'
                        new_file_path = os.path.join(target_dir, new_file_name)
                        # Convert and save the image as .png to the target directory
                        img.save(new_file_path)
                        print(f"Converted and saved: {new_file_path}")
                except IOError:
                    print(f"Cannot convert {file_path}.")

# Specify the path to your source dataset folder here
source_folder = 'C:\\Users\\Alexandra\\Documents\\SPRING_2024\\BMEN_689\\Project_2\\Jupyter_Notebook\\Fixed_Phages_EMs\\'
# Specify the path to your target dataset folder here
target_folder = 'C:\\Users\\Alexandra\\Documents\\SPRING_2024\\BMEN_689\\Project_2\\Jupyter_Notebook\\PNG_Phage_EMs\\'
convert_images_to_png(source_folder, target_folder)
```
